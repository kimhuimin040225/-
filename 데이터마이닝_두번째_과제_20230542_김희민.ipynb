```python
#문제1
import pandas as pd  # pandas 라이브러리 불러오기 (데이터 분석용)

# CSV 파일을 웹에서 읽어와 DataFrame으로 저장
# df = pd.read_csv("members.csv")  # 로컬 파일 사용 시
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

# 1) 전체 데이터의 80% 위치 계산
line = int(len(df) * 0.8)  # 데이터 길이 * 0.8 → 앞에서부터 80% 행 수

# 앞에서부터 80% 데이터만 슬라이싱하여 사용
df = df.iloc[:line]

# 2) 결측치 채우기 전 f1 컬럼의 표준편차 계산
std1 = df['f1'].std()  # NaN 제외하고 표준편차 계산됨

# 3) f1 컬럼의 중앙값 계산
med = df['f1'].median()

# f1 컬럼의 결측치(NaN)를 중앙값으로 채우기
df['f1'] = df['f1'].fillna(med)

# 4) 결측치를 채운 후 f1 컬럼의 표준편차 계산
std2 = df['f1'].std()

# 5) 결측치 처리 전후 표준편차 차이의 절대값 계산
result = abs(std1 - std2)

# 결과를 소수 둘째자리까지 반올림하여 출력
print(round(result, 2))



#문제 2
import pandas as pd  # pandas 라이브러리 불러오기

# CSV 파일을 웹에서 읽어와 DataFrame으로 저장
# df = pd.read_csv("members.csv")  # 로컬 파일 사용 시
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

# 1) age 컬럼의 표준편차 계산
std = df['age'].std()

# age 컬럼의 평균값 계산
mean = df['age'].mean()

# 2) 이상치 기준 하한 계산
# 평균 - (표준편차 × 1.5)
lower = mean - (std * 1.5)

# 이상치 기준 상한 계산
# 평균 + (표준편차 × 1.5)
upper = mean + (std * 1.5)

# 3) 하한보다 작은 값 → 이상치 조건
cond1 = df['age'] < lower

# 상한보다 큰 값 → 이상치 조건
cond2 = df['age'] > upper

# 4) 두 조건 중 하나라도 만족하는 이상치 데이터 선택 후 age 값 합계 계산
print(df[cond1 | cond2]['age'].sum())
```
