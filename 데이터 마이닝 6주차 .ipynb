```python
# 1. 라이브러리 불러오기
import pandas as pd

# 2. 데이터 불러오기
# 로컬 파일 사용 시
# train = pd.read_csv("score_train.csv")
# test = pd.read_csv("score_test.csv")

# 깃허브 주소 사용 시
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch7/score_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch7/score_test.csv")


# 3. 데이터 크기 확인
print("===== 데이터 크기 =====")
print("Train Shape:", train.shape)   # train 데이터의 행, 열 크기 확인
print("Test Shape:", test.shape)     # test 데이터의 행, 열 크기 확인
print("\n")


# 4. 컬럼별 자료형 확인
print("===== 데이터 정보(자료형) =====")
train.info()   # 각 컬럼의 자료형, 결측치 여부 확인
print("\n")


# 5. 결측치 확인
print("===== train 결측치 수 =====")
print(train.isnull().sum().sum())   # train 전체 결측치 개수
print("\n")

print("===== test 결측치 수 =====")
print(test.isnull().sum().sum())    # test 전체 결측치 개수
print("\n")


# 6. target 값 분포 확인
print("===== target 빈도 =====")
print(train['Credit_Score'].value_counts())   # 타겟 클래스 개수 확인
print("\n")


# 7. target 분리
target = train.pop('Credit_Score')
# train에서 Credit_Score 컬럼을 제거하고 target 변수에 저장


# 8. 원핫인코딩
train = pd.get_dummies(train)
test = pd.get_dummies(test)

# train과 test의 컬럼을 맞춰줌
train, test = train.align(test, join='left', axis=1, fill_value=0)


# 9. 검증 데이터 분할
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(
    train,
    target,
    test_size=0.2,
    random_state=0,
    stratify=target
)

print("===== 분할된 데이터 크기 =====")
print("X_tr:", X_tr.shape)
print("X_val:", X_val.shape)
print("y_tr:", y_tr.shape)
print("y_val:", y_val.shape)
print("\n")


# 10. 모델 학습
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr, y_tr)


# 11. 검증 데이터 예측
pred_val = rf.predict(X_val)


# 12. 성능 평가
from sklearn.metrics import f1_score

f1 = f1_score(y_val, pred_val, average='macro')
print("f1-macro:", f1)
print("\n")


# 13. 테스트 데이터 예측
pred_test = rf.predict(test)


# 14. 제출 파일 생성
submit = pd.DataFrame({'pred': pred_test})
submit.to_csv("result.csv", index=False)


# 15. 제출 파일 확인
print("===== 제출파일 상위 5개 =====")
print(pd.read_csv("result.csv").head())
```
